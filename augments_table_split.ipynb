```python
import pandas as pd
```

```python
df = pd.read_csv('/content/game_results_of_each_participants.csv', encoding='EUC-KR')
```

```python
df.head()
```

```python
df['augments'] = df['augments'].str.replace('[','')
df['augments'] = df['augments'].str.replace(']','')
df['augments'] = df['augments'].str.replace("'",'')
df
```

```python
df['augments1'] = df.augments.str.split(', ').str[0]
df['augments2'] = df.augments.str.split(', ').str[1]
df['augments3'] = df.augments.str.split(', ').str[2]

df
```

```python
df = df.drop('augments', axis=1)
df = df.drop('traits', axis=1)
df = df.drop('units', axis=1)
df
```

```python
df.to_csv('augments_table.csv', header=True, index=False, encoding='utf8')
```
